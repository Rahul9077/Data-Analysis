# Bar Chart

```python
x=[1,2,3,4,5]
y=[10,22,33,44,55]
```

```python
plt.bar(x,y)
plt.title('our first bar chart')
plt.xlabel('some x values')
plt.ylabel('some y values')
```

```python
x1=[1,2,3,4,5]
y1=[10,22,33,44,55]
x2=[1,2,3,4,5]
y2=[11,34,56,19,10]
```

```python
plt.figure(figsize=(10,7))
plt.bar(x1,y1,width = .9,label='first chart')
plt.bar(x2,y2,width = .6,label='second chart')
plt.title('Bar Chart')
plt.xlabel('X values')
plt.ylabel('y values')
plt.legend()
```

# Histogram

```python
height = np.arange(150,200,2)
height2 = np.arange(150,200,3)
```

```python
height
```

```python
plt.figure(figsize=(10,8))
plt.hist([height,height2],rwidth=.9,label=['height1','height2'])
plt.legend()
plt.title('Histogram')
plt.xlabel('X Values')
plt.ylabel('Y Values')
```

# Stack Chart/ Area Chart

```python
week = [1,2,3,4,5]
shift1 = [40,35,30,28,33]
shift2 = [40,29,20,37,19]
shift3 = [40,15,19,33,36]
```

```python
plt.figure(figsize=(10,5))
plt.stackplot(week,shift1,shift2,shift3,labels=['shift1','shift2','shift3'])
plt.legend()
```
